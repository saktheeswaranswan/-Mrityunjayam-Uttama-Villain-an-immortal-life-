```python
def nth_digit_of_sqrt2(n):
    """
    Calculate the nth digit of the square root of 2 using the Babylonian method.

    Parameters:
        n (int): The index of the digit to find (0-based).

    Returns:
        int: The nth digit of the square root of 2 (0-9).
    """
    def sqrt_babylonian(number, iterations=50):
        guess = number
        for _ in range(iterations):
            guess = 0.5 * (guess + number / guess)
        return guess

    # Calculate the square root of 2 using the Babylonian method
    sqrt_2 = sqrt_babylonian(2)

    # Shift the number to the left of the decimal point and extract the desired digit
    shifted_number = int(sqrt_2 * 10**n)
    nth_digit = shifted_number % 10

    return nth_digit

# Test the function
n = 10  # Change this value to get the nth digit you desire
result = nth_digit_of_sqrt2(n)
print(f"The {n}th digit of the square root of 2 is {result}.")
```
